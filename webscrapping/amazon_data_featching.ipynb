```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
url='https://www.amazon.in/s?k=android+mobiles&crid=1KVK0N95K97YF&sprefix=android+mobil%2Caps%2C433&ref=nb_sb_noss_2'
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Safari/537.36 Edg/145.0.0.0'}
```

```python
data=[]
```

```python
responce=requests.get(url,headers=headers)
responce.status_code
```

```python
for page in range(1,5):
    params={'k':'mobile','page':page}
    responce=requests.get(url,headers=headers,params=params)
    soup=BeautifulSoup(responce.content,'html.parser')
    products=soup.find_all('div',{'data-component-type':'s-search-result'})
    data.extend(products)
    print(responce.status_code)
```

```python
data1=[]
prices=[]
```

```python
for product in data:
    title_tag=product.h2
    data1.append(title_tag.text.strip())
    product_prices=product.find('span','a-price-whole')
    if product_prices:
        prices.append(product_prices.text.strip())
    else:
        prices.append('N/A')
data1
```

```python
prices
```

```python
brand=[]
for i in data1:
    brand_name=i.split()[0]
    brand.append(brand_name)
brand
```

```python
RAM = []
for i in data1:
    ram = re.search(r'(?i)\b(?:RAM\s*[:\-]?\s*)?(\d{1,2})\s*GB(?:\s*RAM)?(?=[\s,)\]]|$)', i)
    if ram:
        RAM.append(f"{ram.group(1)}GB")
    else:
        RAM.append("N/A")
RAM
```

```python
#rating
ratings=[]
for product in data:
    rating_tag=product.find('span','a-icon-alt')
    if rating_tag:
        ratings.append(rating_tag.text.strip()[:3])
    else:
        ratings.append('N/A')
ratings
```

```python
color=[]
for i in data1:
    color_name=re.search(r'\b(Matte Sliver|Mint Breeze|Regatta|Sand Storm|Tapestry|Black|White|Blue|Red|Green|Yellow|Pink|Purple|Grey|Silver|Gold|UltraMarine|Tendril|Navy|Charcoal|orange|Dune|Titanium)\b',i,re.IGNORECASE)
    if color_name:
        color.append(color_name[0])
    else:
        color.append('N/A')
color
```

```python
storage=[]
for product in data:
    title_tag = product.h2
    title_text = title_tag.get_text(" ", strip=True) if title_tag else ""
    storage_capacity = re.search(r'(?i)\b(?:32|64|128|256|512|1024|2048|2)\s*(?:GB|TB)\b(?:\s*(?:Storage|SSD|HDD|ROM))?', title_text)

    if storage_capacity:
        storage.append(storage_capacity.group(0))
    else:
        storage.append('N/A')
storage
```

```python
cleaned = [
    f"{m.group(1)}{m.group(2).upper()}" if (m := re.search(r'(?i)\b(\d+)\s*(GB|TB)\b', s)) else s
    for s in storage
]

cleaned
```

```python
import re

camera_specs = []
pattern = re.compile(
    r'(?i)'
    r'(?:\b\d{1,3}\s*MP(?:\s*\+\s*\d{1,3}\s*MP)*\b'                       # MP specs
    r'|\b(?:dual|triple|quad|penta|ai|rear|front|selfie|ultra-?wide|wide|'
    r'macro|depth|telephoto|primary|secondary|pro|fusion)\b[\w\s-]{0,30}'
    r'\bcamera(?:\s*system)?\b)'                                          # camera phrases
)

for text in data1:
    matches = [m.group(0).strip() for m in pattern.finditer(text)]
    camera_specs.append(" | ".join(matches) if matches else "N/A")

camera_specs
```

```python
import re

battery_specs = []
pattern = re.compile(r'(?i)\b(\d{4,5})\s*mAh\b(?:\s*battery)?')

for text in data1:
    m = pattern.search(text)
    battery_specs.append(f"{m.group(1)}mAh" if m else "N/A")

battery_specs
```

```python
import re

refresh_rates = []
pattern = re.compile(r'(?i)\b(\d{2,3})\s*hz\b')

for text in data1:
    m = pattern.search(text)
    refresh_rates.append(f"{m.group(1)}Hz" if m else "N/A")

refresh_rates
```

```python
for i in range(len(data1)):
    print("Title",data1[i])
    print("Price :",prices[i])
    print("Brand :",brand[i])
    print("RAM :",RAM[i])
    print("Rating :",ratings[i])
    print("Color :",color[i])
    print("Storage :",cleaned[i])
    print("Camera Specs :",camera_specs[i])
    print("Battery Specs :",battery_specs[i])
    print("Refresh Rate :",refresh_rates[i])
    print("-"*100)
```

```python
df=pd.DataFrame({'Title':data1,'Price':prices,'Brand':brand,'RAM':RAM,'Rating':ratings,'Color':color,'Storage':cleaned,'Camera Specs':camera_specs,'Battery Specs':battery_specs,'Refresh Rate':refresh_rates})
```
