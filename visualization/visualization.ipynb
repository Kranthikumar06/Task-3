```python
import pandas as pd
df=pd.read_csv('Practice-Datasets .csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20,10))
plt.hist(df['Age'], bins=20, color='purple', edgecolor='black', alpha=0.7)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.pie(df['Gender'].value_counts(), labels=df['Gender'].value_counts().index, autopct='%1.1f%%', colors=['lightblue', 'lightpink'])
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.bar(df['City'],df['Customer_ID'],color='orange',alpha=0.5)
plt.title('City Distribution')
plt.xlabel('City')
plt.ylabel('Customer ID')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(x='Monthly_Income', data=df, color='lightgreen')
plt.title('Box Plot of Monthly Income')
plt.xlabel('Monthly Income')
plt.show()
```

```python
#Create a histogram of Spending Score and interpret the skewness
plt.figure(figsize=(20,10))
plt.hist(df['Spending_Score'], bins=20, color='purple', edgecolor='black', alpha=0.4)
plt.title('Distribution of Spending Score')
plt.xlabel('Spending Score')
plt.ylabel('Frequency')
plt.show()
```

```python
#Plot a line chart to visualize Purchase Amount trend over Purchase Date
plt.figure(figsize=(20,10))
plt.plot(df['Purchase_Date'], df['Purchase_Amount'], marker='o', linestyle='-', color='blue')
plt.title('Purchase Amount Trend Over Time')
plt.xlabel('Purchase Date')
plt.ylabel('Purchase Amount')
plt.show()
```

```python
# Create a bar plot to compare average Monthly Income by Gender
plt.figure(figsize=(20,10))
sns.barplot(x='Gender', y='Monthly_Income', data=df,color='lightcoral',alpha=0.6,estimator='mean')
plt.title('Average Monthly Income by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Monthly Income')
plt.show()
```

```python
# Plot a box plot of Purchase Amount grouped by Product Category
plt.figure(figsize=(20,10))
sns.boxplot(x='Product_Category', y='Purchase_Amount', data=df ,palette='Set2')
plt.title('Box Plot of Purchase Amount by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Purchase Amount')
plt.show()
```

```python
#Create a pie chart to show Product Category distribution
plt.figure(figsize=(10,10))
plt.pie(df['Product_Category'].value_counts(), labels=df['Product_Category'].value_counts().index, autopct='%1.1f%%')
plt.title('Product Category Distribution')
plt.axis('equal')
plt.show()
```

```python
#Plot a histogram of Purchase Amount with appropriate bin size
plt.figure(figsize=(20,10))
plt.hist(df['Purchase_Amount'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Purchase Amount')
plt.xlabel('Purchase Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
#Create a bar chart showing average Spending Score by City
plt.figure(figsize=(20,10))
sns.barplot(x='City',y='Spending_Score',data=df,color='magenta',alpha=0.6,estimator='mean')
plt.title('Average Spending Score by City')
plt.xlabel('City')
plt.ylabel('Spending Score')
plt.show()
```

```python
# Plot a line chart showing total Purchase Amount per month
plt.figure(figsize=(20,10))
df['Purchase_Month'] = pd.to_datetime(df['Purchase_Date']).dt.month
monthly_purchase = df.groupby('Purchase_Month')['Purchase_Amount'].sum().reset_index()
plt.plot(monthly_purchase['Purchase_Month'], monthly_purchase['Purchase_Amount'], marker='o', linestyle='-', color='green')
plt.title('Total Purchase Amount Per Month')
plt.xlabel('Month')
plt.ylabel('Total Purchase Amount')
plt.xticks(monthly_purchase['Purchase_Month'])
plt.show()
```

```python
#Create a box plot comparing Monthly Income across different Cities
plt.figure(figsize=(20,10))
sns.boxplot(x='City', y='Monthly_Income', data=df, palette='Set3')
plt.title('Box Plot of Monthly Income by City')
plt.xlabel('City')
plt.ylabel('Monthly Income')
plt.show()
```

```python
# Plot a histogram to analyze Age distribution by Gender
plt.figure(figsize=(20,10))
sns.histplot(data=df, x='Age', hue='Gender', multiple='stack', palette='Set1')
plt.title('Age Distribution by Gender')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Create a bar chart showing total Purchase Amount by Product Category
plt.figure(figsize=(20,10))
sns.barplot(x='Product_Category', y='Purchase_Amount', data=df, estimator='sum', color='pink', alpha=0.7)
plt.title('Total Purchase Amount by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Purchase Amount')
plt.show()
```

```python
#Plot a scatter plot between Monthly Income and Spending Score
plt.figure(figsize=(20,10))
sns.scatterplot(x='Monthly_Income', y='Spending_Score', data=df, color='purple', alpha=0.6)
plt.title('Scatter Plot of Monthly Income vs Spending Score')
plt.xlabel('Monthly Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
#Create a line chart to visualize daily Purchase Amount variation
plt.figure(figsize=(20,10))
daily_purchase = df.groupby('Purchase_Date')['Purchase_Amount'].sum().reset_index()
plt.plot(daily_purchase['Purchase_Date'], daily_purchase['Purchase_Amount'], marker='o', linestyle='-', color='red')
plt.title('Daily Purchase Amount Variation')
plt.xlabel('Purchase Date')
plt.ylabel('Purchase Amount')
plt.show()
```

```python
#Plot a box plot of Spending Score grouped by Gender
plt.figure(figsize=(20,10))
sns.boxplot(x='Gender', y='Spending_Score', data=df, palette='Set1')
plt.title('Box Plot of Spending Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Spending Score')        
plt.show()
```

```python
#Create a bar chart showing customer count by Product Category and Gender
plt.figure(figsize=(20,10))
sns.countplot(x='Product_Category', hue='Gender', data=df, palette='Set1')
plt.title('Customer Count by Product Category and Gender')
plt.xlabel('Product Category')  
plt.ylabel('Customer Count')
plt.show()
```

```python
#Analyze the relationship between Age and Purchase Amount using a suitable visualization
plt.figure(figsize=(20,10))
sns.scatterplot(x='Age', y='Purchase_Amount', data=df, color='teal', alpha=0.6)
plt.title('Scatter Plot of Age vs Purchase Amount')
plt.xlabel('Age')
plt.ylabel('Purchase Amount')
plt.show()
```
